Current dataset has a two level column index which helps to query it.
Below you can see selector for a final dataset. It includes:
1. Object Type of Epitope can only be a Linear peptide
2. Organism Name of Epitope is not Homo sapiens
3. Host Name can be: Homo sapiens, Homo sapiens Black or Homo sapiens Caucasian
4. 1st in vivo Process Type is Occurrence of infectious disease

There are duplicates in the resulting dataset, more filters are applied:
1. epitope is not duplicated
2. epitope has Positive Qualitative Measure (check the condition below)
3. epitope is not in Positive dataset and will be kept with Negative Qualitative Measure

```python
%matplotlib inline
import pandas as pd
```

```python
# reading the dataset
df = pd.read_csv('data/tcell_full_v3.csv', header=[0, 1])
```

```python
df.head()
```

```python
# getting names of the first level columns
df.columns.get_level_values(0).unique()
```

```python
# filtering by epitope type, host and disease 
df = df[(df.Epitope['Object Type'] == 'Linear peptide') & 
        (df.Epitope['Organism Name'] != 'Homo sapiens') & 
        (df.Host.Name.isin(['Homo sapiens', 'Homo sapiens Black', 'Homo sapiens Caucasian'])) & 
        (df['1st in vivo Process']['Process Type'] == 'Occurrence of infectious disease')]
```

```python
# deleting duplicates in columns (Epitope.Description, Assay['Qualitative Measure'])
df = df.drop_duplicates(subset=[('Epitope', 'Description'), ('Assay', 'Qualitative Measure')])
```

```python
# filtering duplicates by epitope. Here we have an entire dataset of duplicates with columns 
# Epitope.Description, Assay['Qualitative Measure'] and Assay['Response Frequency']
dups = df[df.duplicated(subset=[('Epitope', 'Description')], 
                        keep=False)][[('Epitope', 'Description'), ('Assay', 'Qualitative Measure'), 
                                      ('Assay', 'Response Frequency')]].sort_values(by=
                                                                                    [('Epitope', 'Description'), 
                                                                                     ('Assay', 'Response Frequency')])
```

```python
# selecting only rows with epitopes that would be kept with Positive Qualitative Measure:
# Response Frequency should be greater than 50
# using != 
positive = dups[(dups.Assay['Qualitative Measure'] != 'Negative') & (dups.Assay['Response Frequency'] > 50)]
```

```python
# here we have duplicates due to different levels of Positive in Qualitative Measure
positive[positive.duplicated(subset=[('Epitope', 'Description')], 
                        keep=False)][[('Epitope', 'Description'), ('Assay', 'Qualitative Measure'), 
                                      ('Assay', 'Response Frequency')]].sort_values(by=
                                                                                    [('Epitope', 'Description'), 
                                                                                     ('Assay', 'Response Frequency')])
```

```python
# let us drop the rest of duplicates. Here we keep the first occurrence
positive = positive.drop_duplicates(subset=[('Epitope', 'Description')])
```

```python
# filtering the final dataset:
# 1. epitope is not duplicated
# 2. epitope has Positive Qualitative Measure
# 3. epitope is not in Positive dataset and will be kept with Negative Qualitative Measure
df = df[(~df.index.isin(dups.index)) |
        (df.index.isin(positive.index)) |
        ((~df.Epitope.Description.isin(positive.Epitope.Description)) & 
         (df.Assay['Qualitative Measure'] == 'Negative'))]
```

```python
# checking for mistakes: no more duplicates in Epitope.Description
len(df.Epitope.Description) - len(df.Epitope.Description.unique())
```
